```python
FFMPEG_BIN = "ffmpeg"
width = 1920
height = 1080
command = [ FFMPEG_BIN,
            '-i', "/home/mong/Desktop/map/map/videos/VID_20150123_173358_Q.mp4",
            '-f', 'image2pipe',
            '-pix_fmt', 'rgb24',
            '-vcodec', 'rawvideo', '-']

pipe = sp.Popen(command, stdout = sp.PIPE, bufsize=10**8)
rL = list()
gL = list()
bL = list()
i = 0
while True:

     # read 1920*1080*3 bytes (= 1 frame)
    raw_image = pipe.stdout.read(width*height*3)
    # transform the byte read into a numpy array
    image =  numpy.fromstring(raw_image, dtype='uint8')
    frameTotalByteSize = width*height*3
    if len(image) < frameTotalByteSize:
        break
    image = image.reshape((width,height,3))
    # throw away the data in the pipe's buffer.
    pipe.stdout.flush()
    totalRed = 0
    totalGreen = 0
    totalBlue = 0
    for col in range(0, width):
        for row in range(0, height):
            totalRed += image[col][row][0]
            totalGreen += image[col][row][1]
            totalBlue += image[col][row][2]
    r = totalRed/1.0/width/height
    g = totalGreen/1.0/width/height
    b = totalBlue/1.0/width/height
    if not r:
        break
    else:
        rL.append(r)
        gL.append(g)
        bL.append(b)
    i+=1
    print (i, r, g, b)
print("beginning writing to file")
```

```python
writeToFile("/home/mong/Desktop/map/map/videos/VID_20150123_173358_Q.mp4"+".txt", rL, gL, bL)
print ('finished')
```
